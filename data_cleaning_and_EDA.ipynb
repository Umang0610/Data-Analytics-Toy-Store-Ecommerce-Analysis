```python
# CELL A: Load data first
import pandas as pd
df_website=pd.read_csv(r'D:\\data analytics\\Toy Ecommerce dataset\\Maven+Fuzzy+Factory\\cleaned_website_sessions.csv')
df_webpage=pd.read_csv(r'D:\\data analytics\\Toy Ecommerce dataset\\Maven+Fuzzy+Factory\\cleaned_website_pageviews.csv')
df_orders=pd.read_csv(r'D:\\data analytics\\Toy Ecommerce dataset\\Maven+Fuzzy+Factory\\orders.csv')
df_order_items=pd.read_csv(r'D:\\data analytics\\Toy Ecommerce dataset\\Maven+Fuzzy+Factory\\order_items.csv')
df_refunds=pd.read_csv(r'D:\\data analytics\\Toy Ecommerce dataset\\Maven+Fuzzy+Factory\\order_item_refunds.csv')
df_products=pd.read_csv(r'D:\\data analytics\\Toy Ecommerce dataset\\Maven+Fuzzy+Factory\\products.csv')

# CELL B: CONVERT DATES FIRST - BEFORE any other operations
# Different formats need different handling

# Format 1: YYYY-MM-DD HH:MM:SS (with seconds)
for df in [df_website, df_webpage, df_order_items]:
    if 'created_at' in df.columns:
        df['created_at'] = pd.to_datetime(df['created_at'], format='%Y-%m-%d %H:%M:%S')
        print(f"Converted {df.__name__ if hasattr(df, '__name__') else 'df'}")

# Format 2: DD-MM-YYYY HH:MM (no seconds) - THIS IS YOUR ORDERS TABLE
df_orders['created_at'] = pd.to_datetime(df_orders['created_at'], format='%d-%m-%Y %H:%M')
print("Converted orders")

# Format 3: DD-MM-YYYY HH:MM (products table)
df_products['created_at'] = pd.to_datetime(df_products['created_at'], format='%d-%m-%Y %H:%M')
print("Converted products")

# Format 4: DD-MM-YYYY HH:MM (refunds)
df_refunds['created_at'] = pd.to_datetime(df_refunds['created_at'], format='%d-%m-%Y %H:%M')
print("Converted refunds")

# CELL C: Verify conversion worked
print("\n=== VERIFY CONVERSION ===")
for name, df in [('orders', df_orders), ('website', df_website), ('webpage', df_webpage)]:
    print(f"\n{name.upper()} - first date:")
    print(f"  Original value: {df['created_at'].iloc[0]}")
    print(f"  Data type: {df['created_at'].dtype}")
    print(f"  Year extracted: {df['created_at'].dt.year.iloc[0]}")

# CELL D: Now do your EDA (value_counts, etc.)
print("\n=== ORDERS DATE RANGE ===")
print(f"From: {df_orders['created_at'].min()}")
print(f"To: {df_orders['created_at'].max()}")

# CELL E: Then add time components
data_cols = [df_order_items, df_orders, df_webpage, df_website, df_refunds, df_products]
for df in data_cols:
    if 'created_at' in df.columns:
        df['year'] = df['created_at'].dt.year
        df['month'] = df['created_at'].dt.month
        df['day'] = df['created_at'].dt.day
        df['hour'] = df['created_at'].dt.hour
        df['day_of_week'] = df['created_at'].dt.dayofweek

# CELL F: Final verification
print("\n=== FINAL CHECK ===")
print(df_orders[['created_at', 'year', 'month', 'day', 'hour']].head(3))
```
